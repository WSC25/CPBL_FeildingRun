```python
excel_file_path = 'C:/Users/ASUS/OneDrive/Documents/實習/防守數據/計算守備局數/SS守備局數.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name1 ='Sheet1'
```

```python
pos = pd.read_excel(excel_file_path, sheet_name=sheet_name1)
```

```python
pos
```

```python
#輸入聯盟刺殺數和三振數
lg_po = 16025
lg_k = 3866
```

## 計算各守位AVG(POS,聯盟) *各守位公式不同

```python
#投手 二壘 三壘 遊擊 通用
avg_ss = 0.2*(pos['刺殺'].sum()
              +2*pos['助殺'].sum()
              -pos['失誤'].sum()
              +pos['DP'].sum())/(lg_po - lg_k)
avg_ss
```

```python
#一壘手專用
avg_1b = 0.4705*(2*pos['助殺'].sum()
              -pos['失誤'].sum())/(lg_po - lg_k)
avg_1b
```

```python
#捕手專用
avg_c = 0.4705*(pos['刺殺'].sum()
             -pos['SO'].sum()
             +0.4*(pos['助殺'].sum()-pos['CS'].sum())
             -pos['失誤'].sum()
             +pos['DP'].sum()
             +(pos['PB'].sum()/2))/(lg_po - lg_k)
avg_c
```

```python
#外野手專用
avg_of = 0.4705*(pos['刺殺'].sum()
              +4*pos['助殺'].sum()
              -pos['失誤'].sum()
              +pos['DP'].sum())/(lg_po - lg_k)
avg_of
```

## 輸入球隊守備局數和PO-K的值

```python
#輸入球隊的守備局數(投球局數)
IP = {
    "味全龍": 1061.333,
    "統一7-ELEVEn獅": 1061.333,
    "富邦悍將": 1059.333,
    "樂天桃猿": 1081.667,
    "中信兄弟":1078
}

pos["Team_IP"] = pos["球隊"].map(IP)

pos
```

```python
#輸入球隊的PO-K
PO_K = {
    "味全龍": 2414,
    "統一7-ELEVEn獅": 2427,
    "富邦悍將": 2440,
    "樂天桃猿": 2496,
    "中信兄弟":2382
}

pos["team_PO-K"] = pos["球隊"].map(PO_K)

pos
```

## 計算FR  *各守位公式不同

```python
#投手 二壘 三壘 遊擊 通用
pos['FR']=0.2*( pos['刺殺']
              +2*pos['助殺']
              -pos['失誤']
              +pos['DP']) -(pos["team_PO-K"]*avg_ss*(pos["IP"]/pos["Team_IP"]))
```

```python
#一壘手專用
pos['FR']=0.4705*(2*pos['助殺']-pos['失誤'])-(pos["team_PO-K"]*avg_1b*(pos["IP"]/pos["Team_IP"]))
```

```python
#捕手專用
pos['FR'] = 0.4705*(pos['刺殺']
             -pos['SO']
             +0.4*(pos['助殺']-pos['CS'])
             -pos['失誤']
             +pos['DP']
             +(pos['PB']/2))-(pos["team_PO-K"]*avg_c*(pos["IP"]/pos["Team_IP"]))
```

```python
#外野手專用
pos['FR']=0.4705*( pos['刺殺']
              +4*pos['助殺']
              -pos['失誤']
              +pos['DP']) -(pos["team_PO-K"]*avg_of*(pos["IP"]/pos["Team_IP"]))
```
